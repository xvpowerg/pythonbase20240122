## Sort
+ list 進行排序有 sorted( ) 與 sort( ) 兩個動作。
+ 兩者差異：
  + sorted()
    + 不會影響 list 本身結構，且可以輸出結果。
    + sorted(list項目)。
    + sorted( ) 可用來排序任何項目。
+ sort()
    + 會影響 list 本身結構，且不可以於操作這個動作時進行輸出。
    + list.sort( )。
    + sort( ) 只能用在list上排序。


### 排序原則與參數
+ 排序原則：
  + 數字則依照大小排序。
  + 字串則依照 ASCII 編碼順序進行排序。
+ 可加入的參數：
   + reverse=True  
     + 有這個參數由大到小排序
     + 若沒有這個參數則由小到大排序

```python
list4 = [8,9,2,7,6]#預設排序小到大
print(list4)
list4.sort()
print(list4)
list4.sort(reverse = True)#排序大到小
print(list4)
```

```python
# 小寫 > 大寫 > 數字
# a 大於 ZHIJK
# ab 小於  as
# aaa 大於 aa
list5 = ["c","Bz","a"]
print(list5)
list5.sort()
print(list5)
```

## key 與 value
+ key 不能於程式內改變：

  + 不可以使用 list。
+ 如何找出所有的 key 與 value？
  + 您可以透過 dict1.keys( ) 這個方法找出所有的 key。
  + 您可以透過 dict1.values( ) 這個方法找出所有的 value。

```python
dict1 = {'a':100,'b':200,'c':300}
print(dict1.keys())
print(dict1.values())
#幫我輪巡'a', 'b', 'c'
for k in dict1.keys():
    print(k,dict1[k])
```

## Function

  + 減少相同的語法重複出現。

+ 被呼叫的函數必須放在程式區塊的上方，代表已經準備好了，下方的程式區塊才可以呼叫使用。
+ 自訂函數可以放在其他檔案內，之後以模組引用 import 的方式使用：
  + 那自訂函數的內容就交給其他人負責，而你就只負責主程式的流程。
+ 在python函數是一種物件

## 變數影響範圍
+ 函數外的變數：
  + 函數內可以顯示該變數內容
  + 不屬於函數的區域內都可以使用
+ 函數內的變數：
  + 只在函數內產生效果，不會影響函數外的變數
  + 若函數內沒有進行變數宣告而進行改變內容動作將會產生錯誤訊息!

```python
a = 5
def func1():
    a = 10
    print("函數內:",a)
print("函數外1:",a) 
func1()
print("函數外2:",a) 
```

```python
a = 5
def func2():
    global a
    a = 10
    print("函數內:",a)
print("函數外1:",a)   
func2()
print("函數外2:",a)
```

### 不固定接收數量

( ) 內接收資料若是以 * 表示代表可以引入不定數量的參數：

一個*代表以 tuple 的方式引入

兩個*代表以 dict 的方式引入

接收資料若需加入多個設定，這些設定請依照順序排列：

第一個參數設定為接收值 

第二個參數設定為初始值

第三個參數設定為不固定值，可以是 tuple 或 dict 方式

```python
def fun1(age1,age2="default",*age3):
    print(age1)
    print(age2)
    print(age3)
    print("====================")
fun1(1)
fun1(1,2)
fun1(1,2,3)
fun1(1,2,3,4)
```

```python
t1 = (3)
print(type(t1))
t2 = (3,)
print(type(t2))
t3 = (3,2)
print(type(t3))
```

```python
def fun2(age1,age2="default",**arg4):
    print(age1)
    print(age2)
    print(arg4)
    print("====================")
fun2(1)
fun2(1,2)
fun2(1,2,a="b")
fun2(1,2,a="b",name="Ken")
```
