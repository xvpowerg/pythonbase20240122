```python
def fun2(age1,age2="default",**arg4):
    print(age1)
    print(age2)
    print(arg4)
    print("====================")
fun2(1)
fun2(1,2)
fun2(1,2,a="b")
fun2(1,2,a="b",name="Ken")
```
