```python
def mySum(v1=0,*v2):
    total = v1
    for num in v2:
        total += num
    return total
print(mySum())
print(mySum(10))
print(mySum(10,20,30))
```

```python
def myPrint(*agr1,**arg2):
    sep = " "
    if "sep" in arg2:
        sep = arg2["sep"]
        
    for v in agr1:
        print(v,end=sep)

myPrint("A","B","C") 
```

# 匿名函數
+ 不需要定義函數名稱的，只需要用運算式或表達分析語法。
+ Python 使用 lambda 語法定義匿名函數。
+ 匿名函數是一個表達式/計算式，並不是一個執行流程區塊。
+ 匿名函數可以出現在一般函數不允許的地方，例如像 list 內部或函數呼叫參數的位置。

```python
def f(x,y,z):
    return x+y+z
f2 = lambda x,y,z:x+y+z
print(f(2,20,30))
print(f2(2,20,30))
```

```python
f3 = (lambda a = "Test1",b="Test2",c="Test3": a + b + c)
print(f3("v1","v2"))
print(f3("v1"))
```

```python
f4 = [lambda x:x**2,lambda x:x**3,lambda x:x**4]
for f in f4:
    print(f(3))
```

## 匿名函數可用於條件分析表達上

```python
minFun = lambda x,y:y if y<x else x
print(minFun(2,3))
print(minFun(5,1))
```

## Map

```python
prictList = [1,2,3,4]
newList = map(lambda x : x * 2 ,prictList) 
print(newList)
print(list(newList))
```

## Filter

```python
a = list(range(1,7))
print(a)
newList = filter(lambda x:x % 2 == 0,a)
print(newList)
print(list(newList))
```

```python
class Student:
    def __init__(self,name,height=0,weight=0):
        self.name = name
        self.height = height
        self.weight = weight
    
#height與weight預設為0
st1 = Student("Ken")
print(st1.name,st1.height,st1.weight)
#Ken 0 0
st2 = Student("Vivin",165,51)
#Vivin 165 51
print(st2.name,st2.height,st2.weight)
```

```python
class Book:
    def __init__(self,isbn,name):
        self.isbn = isbn
        self.name = name
    def printInfo(self):
        print(self.isbn,self.name)
b1 = Book("A0001","Python AI")        
b2 = Book("A0002","Java Web")  
b3 = Book("A0003","C# .net")  
b1.printInfo()
b2.printInfo()
b3.printInfo()
```

```python
class Student:
    def __init__(self,name,height=0,weight=0):
        self.name = name
        self.height = height
        self.weight = weight
#weight/  (height/100)**2     
    def getBMI(self):
        return self.weight/(self.height/100)**2  
    def printInfo(self):
        bmi = self.getBMI()
        print(self.name,end=" ")
        if bmi < 18.5:
            print("太輕")
        elif bmi <= 25:
            print("正常")
        elif bmi <=30:
            print("過重")
        else:
            print("胖胖") 
#幫我算bmi
st1 = Student("Ken",170,170)
print(st1.getBMI())
st1.printInfo()#Ken 太輕
```
